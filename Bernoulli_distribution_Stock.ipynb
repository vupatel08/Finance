# Bernoulli distribution

https://en.wikipedia.org/wiki/Bernoulli_distribution

```python
import warnings
warnings.filterwarnings("ignore")

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Function to generate Bernoulli trials
def generate_bernoulli_trials(p, size):
    """
    Generate Bernoulli trials with probability p.

    Parameters:
        p (float): Probability of success (1).
        size (int): Number of trials to generate.

    Returns:
        numpy.ndarray: Array of 0s (failure) and 1s (success).
    """
    return bernoulli.rvs(p, size=size)

# Input
symbol = 'AMD'
start = '2014-01-01'
end = '2019-01-01'

# Read data 
dataset = yf.download(symbol, start, end)

# Prepare data
data = dataset['Adj Close']

# Define a probability for success (e.g., 0.5 for a fair coin flip)
p_success = 0.5

# Number of trials (you can adjust this)
num_trials = len(data)

# Generate Bernoulli trials with the defined probability
bernoulli_trials = generate_bernoulli_trials(p_success, num_trials)

# Plot the stock data alongside the Bernoulli trials
plt.figure(figsize=(10, 6))
plt.plot(data.index, data, label='Stock Price')
plt.plot(data.index, bernoulli_trials, label='Bernoulli Trials', linestyle='--', color='red')
plt.title(f'{symbol} Stock Price and Bernoulli Trials')
plt.xlabel('Date')
plt.ylabel('Price / Outcome')
plt.legend()
```

```python
# Plot the stock data in one chart
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(data.index, data, label='Stock Price')
plt.title(f'{symbol} Stock Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

# Plot the Bernoulli trials in a separate chart
plt.subplot(2, 1, 2)
plt.plot(data.index, bernoulli_trials, label='Bernoulli Trials', linestyle='--', color='red')
plt.title('Bernoulli Trials')
plt.xlabel('Date')
plt.ylabel('Outcome')
plt.legend()

plt.tight_layout()  # Ensures proper spacing between subplots

plt.show()
```

```python
# Create a DataFrame from the Bernoulli samples
bernoulli_df = pd.DataFrame({'Date': data.index, 'Bernoulli Samples': bernoulli_trials})

# Print the DataFrame
print(bernoulli_df)
```
